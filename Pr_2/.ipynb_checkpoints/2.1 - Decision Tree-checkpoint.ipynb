# Decision Tree 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.tree import DecisionTreeClassifier, plot_tree
```

## Get the Data 

```python
#https://www.kaggle.com/uciml/iris

url='https://drive.google.com/file/d/10vsX925g8ff4QdgQpnsguo4DHUHOPgAL/view?usp=sharing'
url_='https://drive.google.com/uc?id=' + url.split('/')[-2]
df = pd.read_csv(url_)
```

```python
df.shape
```

```python
df.head()
```

```python
df.columns
```

```python
X = df.drop(['Id', 'Species'], axis = 1)
y = df['Species']
```

```python
set(y) ## labels
```

```python
print(type(X), type(y))
print(X.shape, y.shape)
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

```python
print(type(X), type(y))
print(X_train.shape, y_train.shape)
```

## Decision Tree 

```python
model = DecisionTreeClassifier(criterion='gini')
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

## Model Evaluation

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, y_pred))
```

```python
print(confusion_matrix(y_test, y_pred))
```

## Visualization

```python
plt.figure(figsize = (10, 10))
plot_tree(model, feature_names = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm'], 
          class_names = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'], filled = True)
plt.show()
```
